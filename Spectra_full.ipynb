```python
####### For _282


import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
url = "https://data.sdss.org/sas/dr17/sdss/spectro/redux/26/spectra/lite/1733/spec-1733-53047-0282.fits"
hdul = fits.open(url)
spectrum_data = hdul[1].data
log_wavelength = spectrum_data['loglam']
wavelength = 10 ** log_wavelength
flux = spectrum_data['flux']
emission_lines = {
    'H-alpha': 6562.8,
    'H-beta': 4861.3,
    'O III': 5007.0,
    'He II':1640.4,
    'O III': 1665.85,
    'Al III':1857.4,
    'C III':1908.734,
    'C II':2326.0,
    'Mg II':2799.117, 'Ne V':3346.79,'Ne VI':3426.85,'	O III':5008.240,'O I' :6302.046,'O I':6365.536,"N I":6529.03,"N II":6549.86,"Hα":6564.61,"	N II":6585.27,"S II":6718.29,"S II":6732.67

}

absorption_lines = {
    'Na I D': 5890.0,
    'Ca II H&K': [3968.5, 3933.7],"Mg":5176.7,"Na":5895.6,"CaII":[8500.36,8544.44,8664.52]
}
plt.figure(figsize=(12, 6))
plt.plot(wavelength, flux, label='Spectrum', color='blue')
plt.title('Spectral Data with Marked Lines')
plt.xlabel('Wavelength (Angstroms)')
plt.ylabel('Flux')
for label, angstrom in emission_lines.items():
    plt.axvline(x=angstrom, color='red', linestyle='--', label=f'Emission: {label}')
    plt.text(angstrom, max(flux) * 0.8, label, color='red', rotation=90, verticalalignment='bottom')
for label, angstroms in absorption_lines.items():
    if isinstance(angstroms, list):
        for angstrom in angstroms:
            plt.axvline(x=angstrom, color='green', linestyle='--', label=f'Absorption: {label}')
            plt.text(angstrom, max(flux) * 0.6, label, color='green', rotation=90, verticalalignment='bottom')
    else:
        plt.axvline(x=angstroms, color='green', linestyle='--', label=f'Absorption: {label}')
        plt.text(angstroms, max(flux) * 0.6, label, color='green', rotation=90, verticalalignment='bottom')
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

plt.grid()
plt.show()
```
